```python
import shap
import matplotlib.pyplot as plt

def MLpipe_KFold_RMSE_with_SHAP(X, y, preprocessor, ML_algo, param_grid):
    '''
    Similar to MLpipe_KFold_RMSE but adds SHAP value calculations for local feature importance.

    Returns:
    - test_scores: List of RMSE scores on the test sets.
    - best_models: List of best models from each GridSearchCV run.
    - shap_values_list: List of SHAP values for each test set.
    '''
    test_scores = []
    best_models = []
    shap_values_list = []  # To store SHAP values for each random split

    for random_state in range(10):
        # Split into "other" and test set
        X_other, X_test, y_other, y_test = train_test_split(
            X, y, test_size=0.2, random_state=random_state
        )
        
        # KFold cross-validation
        kfold = KFold(n_splits=4, shuffle=True, random_state=random_state)

        # Define pipeline
        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('model', ML_algo)
        ])
        
        # GridSearchCV for hyperparameter tuning
        grid_search = GridSearchCV(
            pipeline,
            param_grid,
            cv=kfold,
            scoring=make_scorer(root_mean_squared_error, greater_is_better=False),
            n_jobs=-1
        )
        grid_search.fit(X_other, y_other)
        
        # Best model
        best_model = grid_search.best_estimator_
        y_pred_test = best_model.predict(X_test)
        
        # Calculate RMSE on test set
        test_rmse = root_mean_squared_error(y_test, y_pred_test)
        test_scores.append(test_rmse)
        best_models.append(best_model)

        # SHAP explainer for local feature importance
        # Extract the final model from the pipeline
        final_model = best_model.named_steps['model']
        
        # Use SHAP Explainer based on the model type
        explainer = shap.Explainer(final_model, X_other)
        shap_values = explainer(X_test)
        shap_values_list.append(shap_values)
        
        # Optional: Plot SHAP summary plot for this random state
        print(f"Random State {random_state} SHAP Summary Plot:")
        shap.summary_plot(shap_values, X_test, show=False)
        plt.title(f"SHAP Summary Plot - Random State {random_state}")
        plt.show()

    return test_scores, best_models, shap_values_list
```
